```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('Professional Analysis.csv')
```

```python
# Analyse with country, age and gender wise data
#input(country,min_age,max_age,gender)
# displaying (Job title, salary, Race)
```

```python
profession_counts = data['Job Title'].value_counts()[:15]

print(profession_counts)
```

```python

country=input("Enter the country:") .strip()
age_min=input("Enter minimum age:").strip()
age_max=input("Enter maximum age:").strip()
gender=input("Enter the gender:").strip()
df=data
if country.lower()!='all':
    df=df[df['Country'].str.lower()==country.lower()]
if age_min:
    df=df[df['Age']>=int(age_min)]
if age_max:
    df=df[df['Age']<=int(age_max)]
if gender.lower()!='all':
    df=df[df['Gender'].str.lower()==gender.lower()]

plt.figure(figsize=(10, 6))
sns.countplot(y='Job Title',data=df, order=df['Job Title'].value_counts().head(15).index,palette='pastel')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.title(f"Top 15 Job {country.upper()}")
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=20, kde=True, color='#663399')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Race', hue='Country', data=df, palette='pastel', 
                order=df['Race'].value_counts().head(5).index)
plt.title(f'Race wise profession in {country.upper()}')
plt.xlabel('Race')
plt.ylabel('Count')
plt.legend(title='Race')
plt.tight_layout()
plt.show()
```

```python
# Profession analysis with country and experience 
# input(country,min_exp,max_exp)
# Displying(Job Title) 
```

```python
country=input("Enter the country:") .strip()
exp_min=input("Enter minimum Experience:").strip()
exp_max=input("Enter maximum Experience:").strip()
df=data
if country.lower()!='all':
    df=df[df['Country'].str.lower()==country.lower()]
if exp_min:
    df=df[df['Years of Experience']>=int(exp_min)]
if exp_max:
    df=df[df['Years of Experience']<=int(exp_max)]

plt.figure(figsize=(10, 6))
sns.histplot(df['Job Title'], bins=20, kde=True, color='#663399')
plt.title(f'Job opportuniety in {country.upper()} with experience {exp_min}-{exp_max} years')
plt.xlabel('Job Title')
plt.ylabel('Experience')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df=data.groupby('Job Title')['Salary'].mean()
top_5=df.nlargest(5)
plt.figure(figsize=(7,7))
top_5.plot(kind='pie',autopct='%1.2f%%',startangle=90,colors=plt.cm.Paired.colors)
plt.title('Top 5 Job by Average Salary')
plt.ylabel('')
plt.xlabel('Thank You')
plt.show()
```

```python
df=data
job = df['Job Title'].value_counts().sort_values(ascending=False)[:10]
values = job.values
prof = job.index
sns.barplot(x=prof, y=values,color='#0b5345')
plt.title("Top 10 Job In The World")
plt.show()
```
